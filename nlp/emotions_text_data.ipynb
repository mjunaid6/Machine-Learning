```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.txt', sep = ';', names = ['text', 'emotion'])
```

```python
df.isnull().sum()
```

```python
emotions = df['emotion'].unique()
emotions_val = {}
i = 0
for emo in emotions :
    emotions_val[emo] = i
    i += 1

emotions_val
```

```python
df['emotion'] = df['emotion'].map(emotions_val)
```

```python
# 1. Making all text lowercase
df['text'] = df['text'].apply(lambda x : x.lower())
```

```python
# 2. Removing punctuations

import string
def remove_punc(txt): 
    return txt.translate(str.maketrans('', '', string.punctuation))

df['text'] = df['text'].apply(remove_punc)
```

```python
# Removing numbers

def remove_nums(txt): 
    mod_txt = ""
    for i in txt: 
        if not i.isdigit():
            mod_txt += i
    return mod_txt

df['text'] = df['text'].apply(remove_nums)
```

```python
# Removing emojis

def remove_nums(txt): 
    mod_txt = ""
    for i in txt: 
        if i.isascii():
            mod_txt += i
    return mod_txt

df['text'] = df['text'].apply(remove_nums)
```

```python
#removing stopwords

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))

def remove_stopwords(txt) :
    txt = txt.split(' ')
    cleaned_text = ""
    for word in txt:
         if not word in stop_words:
             cleaned_text += word+" "
    return cleaned_text

df['text'] = df['text'].apply(remove_stopwords)
```

```python
# Koi fayda nhi

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(df['text'])

print("Vocabulary : ", vectorizer.get_feature_names_out()) 
print("\nBoW matrix : \n", X.toarray()) 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['emotion'], test_size = 0.20, random_state = 42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
bow_vectorizer = CountVectorizer()
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)

nb_model = MultinomialNB()
nb_model.fit(X_train_bow, y_train)
predict_bow = nb_model.predict(X_test_bow)
print("Accuracy of BoW: ",accuracy_score(y_test, predict_bow))
```

```python
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

nb_model.fit(X_train_tfidf, y_train)
predict_tfidf = nb_model.predict(X_test_tfidf)
print("Accuracy of TFIDF: ",accuracy_score(y_test, predict_tfidf))
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

```python
log_model.fit(X_train_bow,y_train)
predict_bow = log_model.predict(X_test_bow)
print("Accuracy of BoW: ",accuracy_score(y_test, predict_bow))
```

```python
log_model.fit(X_train_tfidf,y_train)
predict_tfidf = log_model.predict(X_test_tfidf)
print("Accuracy of BoW: ",accuracy_score(y_test, predict_tfidf))
```
